# Tutorials

This section provides a set of tutorials that are meant to provide an introduction to R3TAM components through application of component models to the Shasta-Keswick-Sacramento River system.

These tutorials are currently being developed and updated, so check back in the future for more updates. 
For now, the tutorials available include:

1. [Getting started](./tutorials/getting_started.md) - A guide to setting up an appropriate Python environment and installing the R3TAM package
2. [Shasta reservoir simulation](./tutorials/shasta_standalone.md) - An explanation of the inputs required for a ResTemp model as applied to Shasta Reservoir and the steps needed to perform a standalone simulation.
3. [Keswick reservoir simulation](./tutorials/keswick_standalone.md) - An explanation of the inputs required for a LongTemp model as applied to Keswick Reservoir and the steps needed to perform a standalone simulation.
